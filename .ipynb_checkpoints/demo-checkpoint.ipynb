```python
import pymongo
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display_html

# conntect to mongodb
client = pymongo.MongoClient(host='localhost', port=27017)
collection = client.job_data.job_data
```

```python
def stat_pay_ind(sourceweb: str, keyword: str, num: int):
    '''Statisticized jobsalary and jonindcat.
    Counted number of industry types and averaged salary.
    
    Args:
        sourceweb: '104人力銀行' or '1111人力銀行'
        keyword: crawled data keyword
        num: limits the number of documents in the result set.
    
    Returns:
        A DataFrame of statistics.
        | _id | avgpay | counts |
        
        _id: industry types
        avgpay: average salary of industry types
        counts: number of counted   
        
    '''
    
    # pipeline of collection
    pipeline = [
        {'$match': {'sourceweb': sourceweb, 
                   'keyword': keyword}}, 
        {'$group': {'_id': '$jobindcat', 
                    'counts':{'$sum': 1}, 
                   'avgpay': {'$avg': '$jobsalary'}
                   }}, 
        {'$sort': {'counts': -1}}, 
        {'$limit':num}, 
    ]
    data =collection.aggregate(pipeline)
    if data ==0:
        raise print('no data found')
    df = pd.DataFrame(data)
    df = df.round({'avgpay': 2})
   
    return df


def display_side_by_side(*args):
    '''display multiple DataFrame
    Args:
        *arg: DataFrame for display
    '''
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
    

def fig_hist_line(df,df2):
    '''figure histogram of counts and line of avepay.
    Arg:
        df: DataFrame return from stat_pay_ind()
        df2: DataFrame return from stat_pay_ind()
    '''
    fig, ax = plt.subplots(figsize=(15,5))
    fig.subplots_adjust(wspace=0.3)
    ax1 = plt.subplot(1,2,1)
    color = 'tab:blue'
    ax1.bar(df['_id'], df['counts'], color=color)
    ax1.set_ylabel('產業',color=color)
    ax1.tick_params(axis='y', labelcolor=color)
    plt.xticks(rotation=90)
    color = 'tab:red'
    ax2 = ax1.twinx()
    ax2.plot(df['_id'], df['avgpay'], color=color)
    ax2.set_ylabel('薪資', color=color)
    ax2.tick_params(axis='y', labelcolor=color)
    fig.legend(['產業', '薪資'])
    
    
    ax3 = plt.subplot(1,2,2)
    color = 'tab:blue'
    ax3.bar(df2['_id'], df2['counts'], color=color)
    ax3.set_ylabel('產業',color=color)
    ax3.tick_params(axis='y', labelcolor=color)
    plt.xticks(rotation=90)
    color = 'tab:red'
    ax4 = ax3.twinx()
    ax4.plot(df2['_id'], df2['avgpay'], color=color)
    ax4.set_ylabel('薪資', color=color)
    ax4.tick_params(axis='y', labelcolor=color)
    
    fig.legend(['產業', '薪資'])
    return fig
```

# Python

```python
keyword = 'python'
num = 15
df = stat_pay_ind('104人力銀行', keyword, num)
df2 = stat_pay_ind('1111人力銀行', keyword, num)
fig = fig_hist_line(df,df2)
display_side_by_side(df,df2)
```

```python
import datetime
```

```python
sourceweb = '104人力銀行'

pipeline = [
    {'$match': {'sourceweb': '104人力銀行', 
               }},
    {'$group': {'_id': '$keyword'}}
]
key_cat = collection.aggregate(pipeline)
key_list = list(key_cat)
df_n=[]
for i in range(len(key_list)):
    
    s = key_list[i]['_id']
    pipeline = [
        {'$match': {'sourceweb': sourceweb, 
                   'keyword': s, 
                   'jobdate': {'$lt': datetime.datetime.now()}, 
                   }},
        {'$group': {'_id': '$jobdate', 
                    'counts':{'$sum': 1}, 
                   }}, 
        {'$sort': {'counts': 1}}, 
    ]
    date_data = collection.aggregate(pipeline)
    df = pd.DataFrame(date_data).set_index('_id')
    df.columns=[s]
    df_n.append(df)

start = datetime.datetime.strptime('2019/01/01','%Y/%m/%d')
end = datetime.datetime.strptime('2019/05/14','%Y/%m/%d')
r = pd.date_range(start=start, end=end)
ind = pd.DataFrame(index=r)
df_n.append(ind)
df_n = pd.concat(df_n, axis=1, join_axes=[ind.index])

fig = df_n.plot(marker='o', markersize=3, figsize=(15,5))
fig.set_xlabel('日期', fontsize=16)
fig.set_ylabel('更新數量', fontsize=16)
fig.set_title(sourceweb, fontsize=16)
```

```python
sourceweb = '104人力銀行'

pipeline = [
    {'$match': {'sourceweb': '104人力銀行', 
               }},
    {'$group': {'_id': '$keyword'}}
]
key_cat = collection.aggregate(pipeline)
key_list = list(key_cat)
df_n=[]
for i in range(len(key_list)):
    
    s = key_list[i]['_id']
    pipeline = [
        {'$match': {'sourceweb': sourceweb, 
                   'keyword': s, 
                   'jobdate': {'$lt': datetime.datetime.now()}, 
                   }},
        {'$group': {'_id': '$jobapply', 
                    'counts':{'$sum': 1}, 
                   }}, 
        {'$sort': {'counts': 1}}, 
    ]
    date_data = collection.aggregate(pipeline)
    df = pd.DataFrame(date_data).set_index('_id')
    df['counts'] = df['counts']/sum(df['counts'])*100
    df.columns=[s]
    df_n.append(df)
    
r = range(1,3)
ind = pd.DataFrame(index=r)
df_n.append(ind)
df_n = pd.concat(df_n, axis=1, join_axes=[ind.index])
df_n.index=['10人以下', '10~30人']

fig = df_n.plot(kind='bar', rot=0)
fig.set_ylabel('%', fontsize=14)
fig.set_title(sourceweb, fontsize=16)
```
