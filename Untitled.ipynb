```python
import pymongo
import pandas as pd
import matplotlib.pyplot as plt
client = pymongo.MongoClient(host='localhost', port=27017)
collection = client.job_data.job_data
# %matplotlib inline
```

```python
def stat_pay_ind(sourceweb: str, keyword: str, num: int):
    '''Statisticized jobsalary and jonindcat.
    Counted number of industry types and averaged salary.
    
    Args:
        sourceweb: '104人力銀行' or '1111人力銀行'
        keyword: crawled data keyword
        num: limits the number of documents in the result set.
    
    Returns:
        A DataFrame of statistics.
        | _id | avgpay | counts |
        
        _id: industry types
        avgpay: average salary of industry types
        counts: number of counted   
        
    '''
    
    # pipeline of collection
    pipeline = [
        {'$match': {'sourceweb': sourceweb, 
                   'keyword': keyword}}, 
        {'$group': {'_id': '$jobindcat', 
                    'counts':{'$sum': 1}, 
                   'avgpay': {'$avg': '$jobsalary'}
                   }}, 
        {'$sort': {'counts': -1}}, 
        {'$limit':num}, 
    ]
    data =collection.aggregate(pipeline)
    if data ==0:
        raise print('no data found')
    df = pd.DataFrame(data)
    df.set_index(['_id'],inplace=True)
#     total = sum(df['counts'])
#     df['counts'] = df['counts'].map(lambda x: x/total*100)
    

    # figure
    ax = df['counts'].plot(fontsize=12, 
                        kind='bar')
    ax.set_ylabel('間', fontsize=14)
    ax.set_xlabel('')
    ax = df['avgpay'].plot(kind='line', 
                           secondary_y=True, 
                           rot=90, 
                           style='r-.', 
                           use_index=False)
    ax.set_ylabel('元', fontsize=14)
    ax.set_xlabel('')
    ax.set_title(f'{keyword}  in  {sourceweb}', fontsize=16)
    width = .5
    plt.xlim([-width, len(df['counts'])-width])
   
    return df, ax
```

```python
from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
keyword = 'python'
num = 15
fig, axs = plt.subplots(2,1,figsize=(18,5))
plt.subplot(1,2,1)
df1, ax1 = stat_pay_ind('104人力銀行', keyword, num)
plt.subplot(1,2,2)
df2, ax2 = stat_pay_ind('1111人力銀行', keyword, num)
plt.subplots_adjust(wspace = 0.25, hspace =0)

display_side_by_side(df1,df2)
```
